```python
import tkinter as tk
from tkinter import ttk
import math

def calculate():
    num1 = float(input1.get())
    num2 = float(input2.get())
    
    if operation.get() == "Add":
        result = num1 + num2
    elif operation.get() == "Subtract":
        result = num1 - num2
    elif operation.get() == "Multiply":
        result = num1 * num2
    elif operation.get() == "Divide":
        if num2 == 0:
            result = "Cannot divide by zero"
        else:
            result = num1 / num2
    elif operation.get() == "HCF":
        result = math.gcd(int(num1), int(num2))
    elif operation.get() == "LCM":
        result = abs(int(num1 * num2)) // math.gcd(int(num1), int(num2))
    elif operation.get() == "Power":
        result = num1 ** num2
    elif operation.get() == "Log":
        if num1 <= 0 or num2 <= 0:
            result = "Invalid input for logarithm"
        else:
            result = math.log(num1, num2)
    else:
        result = "Invalid operation"
    
    result_label.config(text=f"Result: {result}")


root = tk.Tk()
root.title("Calculator")


input1 = ttk.Entry(root)
input2 = ttk.Entry(root)


result_label = ttk.Label(root, text="Result: ")


operation = tk.StringVar()
operation.set("Add") 

add_radio = ttk.Radiobutton(root, text="Add", variable=operation, value="Add")
subtract_radio = ttk.Radiobutton(root, text="Subtract", variable=operation, value="Subtract")
multiply_radio = ttk.Radiobutton(root, text="Multiply", variable=operation, value="Multiply")
divide_radio = ttk.Radiobutton(root, text="Divide", variable=operation, value="Divide")
hcf_radio = ttk.Radiobutton(root, text="HCF", variable=operation, value="HCF")
lcm_radio = ttk.Radiobutton(root, text="LCM", variable=operation, value="LCM")
power_radio = ttk.Radiobutton(root, text="Power", variable=operation, value="Power")
log_radio = ttk.Radiobutton(root, text="Log", variable=operation, value="Log")


calculate_button = ttk.Button(root, text="Calculate", command=calculate)


input1.grid(row=0, column=0, padx=10, pady=10)
input2.grid(row=0, column=1, padx=10, pady=10)
add_radio.grid(row=1, column=0, padx=10, pady=5)
subtract_radio.grid(row=1, column=1, padx=10, pady=5)
multiply_radio.grid(row=2, column=0, padx=10, pady=5)
divide_radio.grid(row=2, column=1, padx=10, pady=5)
hcf_radio.grid(row=3, column=0, padx=10, pady=5)
lcm_radio.grid(row=3, column=1, padx=10, pady=5)
power_radio.grid(row=4, column=0, padx=10, pady=5)
log_radio.grid(row=4, column=1, padx=10, pady=5)
calculate_button.grid(row=5, columnspan=2, padx=10, pady=10)
result_label.grid(row=6, columnspan=2, padx=10, pady=5)


root.mainloop()
```
