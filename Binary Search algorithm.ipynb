```python
import random

try:
    def binary_search(array, n):
        """
        Perform binary search to find the index of an element in a sorted array.
        """
        first, last = 0, len(array) - 1

        while first <= last:
            mid = (first + last) // 2
            if array[mid] == n:
                return mid
            elif array[mid] < n:
                first = mid + 1
            else:
                last = mid - 1

        return -1

    def search_numbers(array, numbers):
        """
        Search for multiple numbers in an array using binary search.
        """
        results = {}  # Dictionary to store search results for each number
        for n in numbers:
            index = binary_search(array, n)  # Use binary search to find the index of the number
            results[n] = index  # Store the index in the results dictionary
        return results

    array_length = 10
    start_number = random.randint(1, 100)
    array = [start_number + 2 * i for i in range(array_length)]

    while True:
        print("Menu:")
        print("1. Search for a number")
        print("2. Quit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            try:
                input_str = input("Enter numbers to search (comma-separated): ")
                numbers_to_search = list(map(int, input_str.split(',')))
                # Convert user input to a list of numbers to search
                search_results = search_numbers(array, numbers_to_search)
                for n, index in search_results.items():
                    if index != -1:
                        print(f"The index of {n} in the array is {index}")
                    else:
                        print(f"{n} is not in the array.")
            except ValueError:
                print("Invalid input. Please enter valid numbers separated by commas.")
        elif choice == 2:
            print("Search ended.")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

except:
    print("Enter valid inputs.")
```
